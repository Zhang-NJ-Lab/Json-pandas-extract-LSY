```python
dataglass=data3[5].astype(str).str.split(':', expand=True)#玻璃种类
dataglass1=dataglass[1].astype(str).str.split('\'', expand=True)
dataglass2=dataglass1[1]
```

```python
dataff=data2[20].astype(str).str.split(':', expand=True)
dataff1=dataff[1]#ff数值提取
dataff2=dataff1.astype(str).str.split('[', expand=True)
dataff3=dataff2[1].astype(str).str.split(']', expand=True)
dataff4=dataff3[0]
```

```python
datapce=data2[29].astype(str).str.split(':', expand=True)
datapce1=datapce[1]#pce数值提取
datapce2=datapce1.astype(str).str.split('[', expand=True)
datapce3=datapce2[1].astype(str).str.split(']', expand=True)
datapce4=datapce3[0]
```
